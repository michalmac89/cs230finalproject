# Compare the Results for Dropout


## Accuracy
- Reference

```python
training_dev_accuracy_reference = pd.read_csv('drive/My Drive/results/training_dev_accuracy_reference.csv', index_col=0)
training_dev_accuracy_reference.tail()
```

```python
ax = training_dev_accuracy_reference.plot(grid=True)
ax.set_xlabel('# epochs')
ax.set_ylabel('Accuracy')
```

```python
print('Training accuracy', round(training_dev_accuracy_reference['training_accuracy'].values[-1],3))
print('Dev accuracy', round(training_dev_accuracy_reference['dev_accuracy'].values[-1],3))
```

- Dropout keep_prob = 0.8

```python
ax = training_dev_accuracy_dropout_02.plot(grid=True)
ax.set_xlabel('# epochs')
ax.set_ylabel('Accuracy')
```

```python
print('Training accuracy', round(training_dev_accuracy_dropout_02['training_accuracy'].values[-1],3))
print('Dev accuracy', round(training_dev_accuracy_dropout_02['dev_accuracy'].values[-1],3))
```
